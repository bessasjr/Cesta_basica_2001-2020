```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lendo o arquivo que contem valores da cesta básica nas capitais brasileiras de todos os meses, entre 2011 e 2020 (Dados do Dieese)

cest = pd.read_excel('Cesta_basica_br.xls')

cest
```

```python
# Modificando o nome nan na linha 0 coluna 0 para Data

cest[['Gasto Mensal - Total da Cesta']] = cest[['Gasto Mensal - Total da Cesta']].fillna('Data')

cest
```

```python
# transformando a primeira linha (que contem os estados) em colunas

cest.columns = cest.iloc[0]
cest
```

```python
# Deletando a primeira linha(já transformada em colunas)

cest = cest.drop(index=[0])
cest
```

```python
# Deletando as 3 ultimas linhas (que não contém mais dados)

cest = cest.iloc[:120]
cest
```

```python
# Descobrindo a média anual do valor da cesta básica por capital e salvando em um dicionário

anos = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']

lista_mes = []
lista_ano = []

dic = {}

for s in cest.columns[1:]:
    for p in anos:
        for n in range(len(cest['Data'])):
            if cest['Data'][n+1][3:] == p:
                if cest[s][n+1] > 0:
                    lista_mes.append(cest[s][n+1])
                if len(lista_mes) > 0:
                    h = round((sum(lista_mes)/len(lista_mes)), 2)
                else:
                    h = 0
        lista_mes.clear()
        lista_ano.append(h)

    dic[s] = lista_ano[0:10]
    lista_ano.clear()

print(dic)
```

```python
# Criando o Dataframe que contem a média anual dos valores da cesta básica nas capitais brasileiras

cesta = pd.DataFrame(dic, index=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])

cesta
```

```python
# Retirando as colunas que contem elementos com valor 0
cesta1 = cesta.loc[[2012]]
for i in cesta.columns:
    if float(cesta1[i]) == 0:
        cesta1 = cesta1.drop(columns=[i])

cesta1
```

```python
# Criando duas funções para descobrir os valores máximos e mínimos da cesta básica em cada ano observado

def maximum(year):
    maior = cesta.loc[year].idxmax()
    return maior

def minimum(year):
    cesta1 = cesta.loc[[year]]
    for i in cesta.columns:
        if float(cesta1[i]) == 0:
            cesta1 = cesta1.drop(columns=[i])
    menor = cesta1.loc[year].idxmin()
    return menor

print(f'{"Data"}  -  {"Cidade mais cara":.<10}{"Valor":>11}     {"Cidade mais barata":.<12}{"Valor":>9}')

for i in cesta.index:
    print(f'{i}  -  {maximum(i):.<18}{cesta[maximum(i)][i]:.>10.2f}    {minimum(i):.<18}{cesta[minimum(i)][i]:.>10.2f}' )
```

```python
# Descobrindo a média anual na soma de todas as capitais brasileiras

dict_media_anual = {}

for p in cesta.index:
    count = 0
    for i in cesta.loc[p]:
        if i > 0:
            count += 1

    media_anual = round(((sum(cesta.loc[p])) / count), 2)
    dict_media_anual[p] = media_anual

dict_media_anual
```

```python
cesta_basica = pd.DataFrame(dict_media_anual.values(), index=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020], columns=['Mean'])

cesta_basica
```
