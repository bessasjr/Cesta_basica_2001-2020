```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lendo o arquivo que contem valores da cesta básica nas capitais brasileiras de todos os meses, entre 2011 e 2020 (Dados do Dieese)

cest = pd.read_excel('Cesta_basica_br.xls')

cest
```

```python
# Modificando o nome nan na linha 0 coluna 0 para Data

cest[['Gasto Mensal - Total da Cesta']] = cest[['Gasto Mensal - Total da Cesta']].fillna('Data')

cest
```

```python
# transformando a primeira linha (que contem os estados) em colunas

cest.columns = cest.iloc[0]
cest
```

```python
# Deletando a primeira linha(já transformada em colunas)

cest = cest.drop(index=[0])
cest
```

```python
# Deletando as 3 ultimas linhas (que não contém mais dados)

cest = cest.iloc[:120]
cest
```

```python
# Descobrindo a média anual do valor da cesta básica por capital e salvando em um dicionário

anos = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']

lista_mes = []
lista_ano = []

dic = {}

for s in cest.columns[1:]:
    for p in anos:
        for n in range(len(cest['Data'])):
            if cest['Data'][n+1][3:] == p:
                if cest[s][n+1] > 0:
                    lista_mes.append(cest[s][n+1])
                if len(lista_mes) > 0:
                    h = round((sum(lista_mes)/len(lista_mes)), 2)
                else:
                    h = 0
        lista_mes.clear()
        lista_ano.append(h)

    dic[s] = lista_ano[0:10]
    lista_ano.clear()

print(dic)
```

```python
# Criando o Dataframe que contem a média anual dos valores da cesta básica nas capitais brasileiras

cesta = pd.DataFrame(dic, index=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])

cesta
```

```python
# Retirando as colunas que contem elementos com valor 0

cesta1 = cesta.loc[[2012]]
for i in cesta.columns:
    if float(cesta1[i]) == 0:
        cesta1 = cesta1.drop(columns=[i])

cesta1
```

```python
# Criando duas funções para descobrir os valores máximos e mínimos da cesta básica em cada ano observado

city_high = []
value_high = []
city_low = []
value_low = []
difference = []

def maximum(year):
    maior = cesta.loc[year].idxmax()
    return maior

def minimum(year):
    cesta1 = cesta.loc[[year]]
    for i in cesta.columns:
        if float(cesta1[i]) == 0:
            cesta1 = cesta1.drop(columns=[i])
    menor = cesta1.loc[year].idxmin()
    return menor

print(f'{"Data"}  -  {"Cidade mais cara":.<10}{"Valor":>11}     {"Cidade mais barata":.<12}{"Valor":>9}')

for i in cesta.index:
    city_high.append(maximum(i))
    value_high.append(cesta[maximum(i)][i])
    city_low.append(minimum(i))
    value_low.append(cesta[minimum(i)][i])
    difference.append(cesta[maximum(i)][i] - cesta[minimum(i)][i])

    print(f'{i}  -  {maximum(i):.<18}{cesta[maximum(i)][i]:.>10.2f}    {minimum(i):.<18}{cesta[minimum(i)][i]:.>10.2f}' )
    

```

```python
# Gráfico com valores anuais da cesta básica da cidade mais cara e mais barata entre 2011 e 2020

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(18,8))

p1 = ax.bar(cesta.index,value_high, label='Cidades mais caras')
p2 = ax.bar(cesta.index,value_low, label='Cidades mais baratas')


ax.bar_label(p1, label_type='edge', fontsize=14)
ax.bar_label(p2, label_type='center', fontsize=14)
ax.legend(fontsize=16)

ax.set_xticks(cesta.index, labels=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])

#plt.text(value_high[0], ha = 'center', bbox = dict(facecolor ='cyan', alpha =0.8))

#plt.xticks(np.arange(min(cesta.index), max(cesta.index)+1, 1.0))

plt.show()
```

```python
# Diferença percentual entre o valor da cesta básica na cidade mais cara e mais barata entre 2011 e 2020

percentual_maior_menor = []

print('Data    Cidade mais cara      Cidade Mais Barata      Diferença percentual')

for i in range(len(value_high)):
    perc = round((((value_high[i] / value_low[i]) -1) * 100),2)
    percentual_maior_menor.append(perc)
    print(f'{cesta.index[i]:<10}{city_high[i]:<25}{city_low[i]:<22}{perc:>8}%')

percentual_maior_menor
```

```python
# Gráfico que mostra a diferença percentual entre o valor da cesta básica na cidade mais cara e mais barata entre 2011 e 2020

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(18,10))

index = [0, 10, 20, 30, 40, 50]

p = ax.barh(cesta.index, percentual_maior_menor, label='Cidades mais caras') #color=['#FFE4B5', '#FFDEAD', '#FFDAB9', '#FFE4C4', '#FFEBCD', '#FFEFD5','#FAEBD7', '#FAF0E6', '#FDF5E6', '#FFFAF0'])

ax.bar_label(p, label_type='edge', fontsize=18, bbox = dict(facecolor = 'cyan', alpha =0.8))

ax.set_yticks(cesta.index, labels=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])
ax.set_xticks(index, labels=[0, 10, 20, 30, 40, 50])

ax.set_title('Diferença percentual entre a cesta básica mais cara e mais barata no período', fontsize=25, weight='semibold', pad=20)

ax.set_xlabel('Valor Percentual', fontsize=20)
ax.set_ylabel('Ano', fontsize=20)

plt.show()
```

```python
# Descobrindo a média anual do valor da cesta básica entre todas as capitais brasileiras entre 2011 e 2020

dict_media_anual = {}

for p in cesta.index:
    count = 0
    for i in cesta.loc[p]:
        if i > 0:
            count += 1

    media_anual = round(((sum(cesta.loc[p])) / count), 2)
    dict_media_anual[p] = media_anual

dict_media_anual
```

```python
# Criando um dataframe com a média anual do valor da cesta básica entre as capitais brasileiras entre 2011 e 2020

cesta_basica = pd.DataFrame(dict_media_anual.values(), index=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020], columns=['Mean'])

cesta_basica
```

```python
# Gráfico que mostra a média anual do valor da cesta básica entre as capitais brasileiras entre 2011 e 2020

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(14,8))

p1 = ax.bar(cesta_basica.index, cesta_basica['Mean'], width=0.5)
p2 = ax.plot(cesta_basica.index, cesta_basica['Mean'], color='r')

ax.bar_label(p1, fontsize=12, padding=10, color='black')

ax.set_xticks(cesta_basica.index, labels=cesta_basica.index)

plt.show()
```
